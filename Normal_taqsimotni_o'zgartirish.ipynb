```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Eksponensial taqsimot
X = np.random.exponential(2, (10000,1))
#Log-normal taqsimot
Y = np.random.weibull(a = 2, size = (10000,1))
```

##Tasvirlash: Exponential va Weibull taqsimotlari

```python
plt.figure(figsize = (10,4), dpi = 120)

#Exponential taqsimotini tasvirlash
plt.subplot(121)
plt.hist(X, bins = 100)
plt.title("exponential distribution")

#Weibull taqsimotini tasvirlash
plt.subplot(122)
plt.hist(Y, bins = 100)
plt.title("weibull distribution")
```

```python
from sklearn.preprocessing import QuantileTransformer
qt = QuantileTransformer(n_quantiles = len(X), output_distribution = 'normal')

#Yuqoridagi taqsimotlarni Normal taqsimotiga o'zgartirish
X = qt.fit_transform(X)
Y = qt.fit_transform(Y)
print('distributions transformed')
```

##O'zgartirilgan taqsimotlarni tasvirlash

```python
plt.figure(figsize = (10,4), dpi = 120)

#O'zgartirilgan Exponential taqsimotini tasvirlash
plt.subplot(121)
plt.hist(X, bins = 100)
plt.title("transformed exponential")

#O'zgartirilgan Weibull taqsimotini tasvirlash
plt.subplot(122)
plt.hist(Y, bins = 100)
plt.title("transformed weibull")
```
